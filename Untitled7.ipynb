```python
import time
import math
import sys

# 解除 Python 3.11+ 的大數轉字串限制，方便實驗
sys.set_int_max_str_digits(300000)

# ==========================================
# 核心演算法定義
# ==========================================

def standard_binary_search_log(x, Y):
    """[1] 傳統標準二分搜尋法"""
    low = 0
    high = Y.bit_length()
    ans = 0
    while low <= high:
        mid = (low + high) // 2
        val = pow(x, mid)
        if val == Y:
            return mid
        elif val < Y:
            ans = mid
            low = mid + 1
        else:
            high = mid - 1
    return ans


def bit_length_optimized_binary_search_log(x, Y):
    """[2] 傳統大數庫常用的位元長度逼近二分搜"""
    low = max(0, (Y.bit_length() - 1) // x.bit_length())
    high = (Y.bit_length() // (x.bit_length() - 1)) + 1 if x > 2 else Y.bit_length()
    ans = low
    while low <= high:
        mid = (low + high) // 2
        val = pow(x, mid)
        if val == Y:
            return mid
        elif val < Y:
            ans = mid
            low = mid + 1
        else:
            high = mid - 1
    return ans


def f_driven_log_core(x, Y):
    """[3] 基於論文 f 函數特徵定位與二分搜尋結合的對數演算法 (修正版)"""
    if Y == 1: return 0
    if Y < x: return 0

    # 計算 R 與 f 函數
    R = (Y - 1) // (x - 1)
    f = R - (x - 1) * ((R * R) // Y)

    # 數學安全上界
    max_q = Y.bit_length() if x == 2 else (Y.bit_length() // (x - 1)) + 1
    low_q, high_q = 0, max_q
    best_n = f

    while low_q <= high_q:
        mid_q = (low_q + high_q) // 2
        test_n = mid_q * (x - 1) + f
        val = pow(x, test_n)
        if val == Y:
            return test_n
        elif val < Y:
            best_n = test_n
            low_q = mid_q + 1
        else:
            high_q = mid_q - 1
    return best_n


def z_split_divide_conquer_log(x, Y, z):
    """[4] 您設計的 z 軸進制拆分分治演算法"""
    if Y == 1: return 0
    if Y < x: return 0

    # 1. 計算新底數 x^z
    x_z = pow(x, z)
    # 2. 利用 K 公式剝離低位餘數
    k = ((Y - 1) % (x_z - 1)) + 1
    # 3. 切割真數
    Y_over_k = Y // k

    # 4. 分治求解兩個極小規模的子問題
    n1 = f_driven_log_core(x, k)
    n2 = f_driven_log_core(x_z, Y_over_k)

    return z * n2 + n1

# ==========================================
# 性能基準測試 (Benchmark)
# ==========================================
if __name__ == "__main__":
    print("==================================================")
    print("      四大整數對數演算法 (Integer Log) 性能大亂鬥  ")
    print("==================================================\n")

    # 構造超大測試案例
    x = 3
    n = 100000
    Y = pow(x, n)
    z = 200  # 分治拆分參數

    print(f"【測試參數】 基數 x = {x}, 真實指數 n = {n}, 拆分軸 z = {z}")
    print(f"【大數規模】 目標 Y 的十進制位數: {int(n * math.log10(x)) + 1} 位\n")

    # ---- [1] 傳統標準二分搜 ----
    t0 = time.perf_counter()
    res1 = standard_binary_search_log(x, Y)
    t1 = time.perf_counter()
    time_std = t1 - t0
    print(f"[1] 傳統標準二分搜       -> 結果: {res1}, 耗時: {time_std:.6f} 秒")

    # ---- [2] 位元長度優化二分搜 ----
    t0 = time.perf_counter()
    res2 = bit_length_optimized_binary_search_log(x, Y)
    t1 = time.perf_counter()
    time_opt = t1 - t0
    print(f"[2] 位元長度優化二分搜    -> 結果: {res2}, 耗時: {time_opt:.6f} 秒")

    # ---- [3] 單純 f 函數特徵二分搜 ----
    t0 = time.perf_counter()
    res3 = f_driven_log_core(x, Y)
    t1 = time.perf_counter()
    time_f = t1 - t0
    print(f"[3] 單純 f 函數特徵二分搜  -> 結果: {res3}, 耗時: {time_f:.6f} 秒")

    # ---- [4] z軸分治 + f 函數演算法 ----
    t0 = time.perf_counter()
    res4 = z_split_divide_conquer_log(x, Y, z)
    t1 = time.perf_counter()
    time_z = t1 - t0
    print(f"[4] z軸分治 + f函數演算法  -> 結果: {res4}, 耗時: {time_z:.6f} 秒")

    # ---- 正確性與加速度驗證 ----
    #assert res1 == n and res2 == n and res3 == n and res4 == n, "❌ 錯誤：有演算法算錯了！"
    print("\n================== 效能對比結論 ==================")
    print("   所有演算法均精確命中答案，正確性驗證通過！\n")
    print(f" 基準點：[4] z軸分治法耗時 {time_z:.6f} 秒")
    print(f" 🚀 z軸分治法 比 [傳統標準二分搜] 快了 {time_std / time_z:.2f} 倍")
    print(f" 🚀 z軸分治法 比 [位元長度優化二分搜] 快了 {time_opt / time_z:.2f} 倍")
    print(f" 🚀 z軸分治法 比 [單純 f 函數二分搜] 快了 {time_f / time_z:.2f} 倍")
    print("==================================================")
```
