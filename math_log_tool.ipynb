```python

def calculate_exact_log():
    print("=== 純整數代數對數還原器 (無傳統 Log 函數) ===")

    try:
        # 1. 讀取輸入
        y = int(input("請輸入數字 (真數 y, 限制 1~2147483648): "))
        x = int(input("請輸入底數 (基底 x, 需大於 1): "))

        # 硬性防呆限制：真數不超過 2147483648
        if y <= 0 or y > 2147483648:
            print("[錯誤] 真數 y 必須在 1 到 100 之間！")
            return
        if x <= 1:
            print("[錯誤] 底數 x 必須大於 1！")
            return

        # 2. 核心代數演算法 (完全不用傳統 log 函數，固定補償指數 z)
        z = 5

        numerator = (y ** z) - 1
        denominator = (x ** z) - 1

        # 檢查是否為完美的整數次方組合
        if numerator % denominator != 0:
            print(f"[提示] 在整數域中，{y} 不是 {x} 的完美整數次方。")
            return

        # 純整數除法 (//)，得到基礎放大項
        base_value = numerator // denominator

        # 計算進位修正項
        big_y = y ** z
        carry_value = (base_value ** 2) // big_y

        # 代數邏輯精確還原答案 n
        n = base_value - (denominator * carry_value)

        # 3. 輸出結果與安全校驗
        if x ** n == y:
            print(f"\n[計算成功] 精確整數解 n = {n}")
            print("--------------------------------------------------")
            print("程式碼特色解析：")
            print("1. 全程『沒有引進任何 math.log 或數學套件』。")
            print("2. 僅透過最基礎的整數乘方(**)與取底除法(//)完成運算。")
        else:
            print("\n[無法對應] 輸入組合在整數域中無解。")

    except ValueError:
        print("[錯誤] 請輸入有效的正整數！")

if __name__ == "__main__":
    calculate_exact_log()
```
